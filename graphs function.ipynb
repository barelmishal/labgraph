# code that make the groups

```python
def makeGroup(df, column,  start, end):
    return [f"{column}_{i}"  for i in range(start, end+1)]

group1 = data[makeGroup(data, "waterupa", 2, 3)]
group2 = data[makeGroup(data, "waterupa", 5, 10)]
group3 = data[makeGroup(data, "waterupa", 10, 16)]
group4 = data[makeGroup(data, "waterupa", 10, 12)]
time = data['Date_Time_1']

Details = {
    "title table": 'Continuous, variable value error bars',
    "yaxis_title": 'pedestrian locomotion (m)'
}
```

# code that make the graphs

```python
def linearRegratioModelGraph(Details, x_axis, *y_axiss):
    color_pla = ['rgb(0, 204, 153)', 'rgb(204, 102, 0)', 'rgb(115, 153, 0)', 'rgb(230, 153, 0)', 'rgb(102, 102, 153)', 'rgb(0, 38, 153)', 'rgb(0, 153, 255)' 'rgb(51, 153, 0)', 'rgb(255, 128, 223)', 'rgb(204, 0, 0)']
    fig = go.Figure([
        go.Scatter(
        name='group_1',
        x=x_axis,
        y=y_axiss[i].mean(axis=1),
        mode='lines',
        line=dict(color=color_pla[i]),
    )
        for i in range(3)
    ])
    fig.update_layout(
        xaxis_title = 'Time',
        yaxis_title=Details["yaxis_title"],
        title=Details["title table"]
    )
    return fig.show()

linearRegratioModelGraph(Details,time,group1,group2,group3)
```
