# linear regration model function

- אפיון הגרף מבחינת נראות

```python
def makeGroup(df, column,  start, end):
    return [f"{column}_{i}"  for i in range(start, end+1)]

group1 = data[makeGroup(data, "vo2", 1, 4)]
group2 = data[makeGroup(data, "vo2", 5, 10)]
group3 = data[makeGroup(data, "vo2", 10, 16)]
group4 = data[makeGroup(data, "vo2", 10, 12)]
time = data['Date_Time_1']

Details = {
    "title table": 'Continuous, variable value error bars',
    "yaxis_title": 'pedestrian locomotion (m)'
}


def linearRegratioModelGraph(Details, x_axis, *y_axiss):
    fig = go.Figure([
    go.Scatter(
        name='group_1',
        x=x_axis,
        y=y_axiss[0].mean(axis=1),
        mode='lines',
        line=dict(color='rgb(31, 119, 180)'),
    ),
    go.Scatter(
        name='group_3',
        x=x_axis,
        y=y_axiss[1].mean(axis=1),
        mode='lines',
        line=dict(color='rgb(45, 201, 25)'),
    ),
    go.Scatter(
        name='group_4',
        x=x_axis,
        y=y_axiss[2].mean(axis=1),
        mode='lines',
        line=dict(color='rgb(80, 100, 46)'),
    )
    ])
    fig.update_layout(
        xaxis_title = 'Time',
        yaxis_title=Details["yaxis_title"],
        title=Details["title table"]
    )
#     fig.update_layout(hovermode="x unified")
    fig.update_layout(hovermode="x")
    return fig.show()

linearRegratioModelGraph(Details,time,group1,group2,group3)
```
