```python
def linearRegratioModelGraph(Details, x_axis, *y_axiss):
    color_pla = ['rgb(0, 204, 153)', 'rgb(204, 102, 0)', 'rgb(115, 153, 0)', 'rgb(230, 153, 0)', 'rgb(102, 102, 153)', 'rgb(0, 38, 153)', 'rgb(0, 153, 255)' 'rgb(51, 153, 0)', 'rgb(255, 128, 223)', 'rgb(204, 0, 0)']
    fig = go.Figure([
        go.Scatter(
        name=f'group_{i+1}',
        x=x_axis,
        y=y_axiss[i].mean(axis=1),
        mode='lines',
        line=dict(color=color_pla[i]),
    )
        for i in range(len(y_axiss))
    ])
    fig.update_layout(
        xaxis_title = 'Time',
        yaxis_title=Details["yaxis_title"],
        title=Details["title table"]
    )
    fig.add_shape(type="rect",
    xref="x", yref="y",
    x0='10/12/2020 12:37:00', y0=0,
    x1='10/12/2020 19:57:00', y1=2.5,
    opacity=0.2,
    fillcolor="orange",
    line_color="orange",)
    
    fig.add_shape(type="rect",
    xref="x", yref="y",
    x0='10/13/2020 07:02:00', y0=0,
    x1='10/13/2020 19:57:00', y1=2.5,
    opacity=0.2,
    fillcolor="orange",
    line_color="orange",)
    fig.add_shape(type="rect",
    xref="x", yref="y",
    x0='10/14/2020 07:02:00', y0=0,
    x1="10/14/2020 18:47:00", y1=2.5,
    opacity=0.2,
    fillcolor="orange",
    line_color="orange",)
    return fig.show()

linearRegratioModelGraph(Details,time,group1,group2,group3,group4)
```
